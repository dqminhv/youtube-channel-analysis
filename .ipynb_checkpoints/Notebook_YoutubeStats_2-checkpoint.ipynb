# Handling Missing Values

```python
df['category'].fillna('Other', inplace=True)
df['channel_type'].fillna('Other', inplace=True)
```

# Analysis

## Top categories

**Let see which categories that those top Youtube channels belongs to**

```python
sns.countplot(x='category', data=df, stat='percent', order=df['category'].value_counts().index,\
              hue='category', legend=False)
plt.xticks(rotation=90)
plt.show()
```

**From this chart, we can see that the most successful categories are about entertainment. The top 5 categories are Entertainment, Music, People & Blog, Gaming and Comedy**

**From this, let's see the total number of subscribers for each category.**

```python
df_subscribers = df.groupby('category', as_index=False).agg({'subscribers':'sum'}).\
                    sort_values('subscribers', ascending=False)
sns.barplot(data=df_subscribers, x='category', y='subscribers', hue='category', legend=False)
plt.xlabel('Chanel Category')
plt.xticks(rotation=90)
plt.ylabel('Number of subscriber')
```

**In terms of numbers of subscribers, Music has a little more subscribers than Entertainment. The order of channel categories seem to be the same.**

## Top countries

**Next, let's see which countries that those top Youtube channels belong to.**

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Country', data=df, order=df['Country'].value_counts().index,\
             stat='percent', hue='Country', legend=False)
plt.xticks(rotation=90)
plt.show()
```

**Let's take population into account.**

```python
df_population = df[['Country', 'Population']].groupby('Country', as_index=False).agg({'Population':'first'})
```

```python
df_channel_by_country = pd.DataFrame(df['Country'].value_counts()).reset_index()
```

```python
df_channel_by_country = pd.merge(df_channel_by_country, df_population, on='Country')
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Country', y='count', size='Population', data=df_channel_by_country, hue='Country', legend=False)
plt.xticks(rotation=90)
plt.show()
```

**The most successful channels mostly reside in the United States, which is about 32%. About half of the most successful channels are in the United States or India. This is quite normal since the United States and India have high populations. Supprisingly, however, China does not have many channels in the top list.**

## Categories in each country

**We examine for each country, what categories are the most successful.**

```python
df_country_category = df[['Country','category']]
```

```python
df_country_category.head()
```

```python
for country in df_country_category['Country'].unique():
    plt.figure(figsize=(8,6))
    sns.countplot(data=df_country_category[df_country_category['Country']==country], x='category',\
                  hue='category', legend=False,\
                  order=df_country_category[df_country_category['Country']==country]['category'].value_counts().index)
    plt.xlabel('Category')
    plt.xticks(rotation=90)
    plt.title('{}'.format(country))
    plt.show()
    plt.close()
```

**Print out the top category for each country**

```python
for country in df_country_category['Country'].unique():
    df = pd.DataFrame(df_country_category[df_country_category['Country']==country]['category'].value_counts())
    # Assuming df is your DataFrame
    first_index = df.index[0]
    first_value = df.iloc[0]
    print('{}'.format(country))
    print("Top Category:", first_index)
    print("Number of channels:", first_value.values)
    print('-----')
```

## How channel's upload rate, channel age, country population, urban population and gross tertiary education enrollment correlate with the number of subscribers of a channel. 

```python
#There are 5 channels that do not have the created year. Let address this problem by searching for the channel
#name in Youtube website and determine its created year
df[df['created_year'].isnull()]
```

```python
df.loc[df['rank']==237, 'created_year'] = 2006
df.loc[df['rank']==469, 'created_year'] = 2008
df.loc[df['rank']==509, 'created_year'] = 2009
df.loc[df['rank']==736, 'created_year'] = 2013
df.loc[df['rank']==763, 'created_year'] = 2017
```

```python
#Create a subset dataframe that includes required columns
df_2 = df[['subscribers','uploads','created_year', 'Gross tertiary education enrollment (%)',
   'Population','Unemployment rate', 'Urban_population','category']]
```

```python
#Create a columns name 'channel_age'
df_2['channel_age'] = 2023 - df_2['created_year']
```

```python
df_2.drop(columns=['created_year'], inplace=True, axis=1)
```

### Correlation Heat Map

```python
#Create a heatmap

corr_matrix = df_2.drop('category', axis=1).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Scatter Plots

```python
for col in df_2.select_dtypes(include='number'):
    plt.figure(figsize=(10, 10))
    sns.relplot(data=df_2, y='subscribers', x=col, hue='category')
    plt.title(f'Number of Subscribers vs {col}')
    plt.show()
```

**From the correlation heat map and the scatter plots, we do not see any significant correlation between upload rate, channel age, country population, urban population and gross tertiary education enrollment and the number of subscribers.**
